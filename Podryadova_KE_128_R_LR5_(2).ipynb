mean() - это функция, которая вычисляет среднее значение переменной. Она принимает в качестве аргумента вектор чисел и возвращает среднее значение этого вектора. Например, mean(c(1, 2, 3, 4, 5)) вернет 3.

median() - это функция, которая вычисляет медиану переменной. Она принимает в качестве аргумента вектор чисел и возвращает медиану этого вектора. Например, median(c(1, 2, 3, 4, 5)) вернет 3.

sd() - это функция, которая вычисляет стандартное отклонение переменной. Она принимает в качестве аргумента вектор чисел и возвращает стандартное отклонение этого вектора. Например, sd(c(1, 2, 3, 4, 5)) вернет 1.58113883008418.

prop.table() - это функция, которая вычисляет вероятности для каждого уникального значения вектора. Она принимает в качестве аргумента вектор чисел и возвращает вектор вероятностей. Например, prop.table(c(1, 2, 3, 4, 5)) вернет вектор c(0.2, 0.4, 0.4, 0.2, 0.2).

var() - это функция, которая вычисляет дисперсию переменной. Она принимает в качестве аргумента вектор чисел и возвращает дисперсию этого вектора. Например, var(c(1, 2, 3, 4, 5)) вернет 2.5.

IQR() - это функция, которая вычисляет интерквартальный размах переменной. Она принимает в качестве аргумента вектор чисел и возвращает интерквартальный размах этого вектора. Например, IQR(c(1, 2, 3, 4, 5)) вернет 1.5.

summary() - это функция, которая выводит сводку статистических показателей для переменной. Она принимает в качестве аргумента вектор чисел и выводит минимальное, максимальное, среднее значение, стандартное отклонение, медиану и интерквартальный размах этого вектора. Например, summary(c(1, 2, 3, 4, 5)) выведет следующее: Min. 1.000 1st Qu. 2.000 Median 3 Mean 3.200 3rd Qu. 4.000 Max. 5.000 NA's 0 Standard Deviation 1.581139

# Основополагающие допущения ANOVA
Как и для многих статистических тестов, существуют некоторые допущения, которым необходимо соответствовать, чтобы иметь возможность интерпретировать результаты. Когда одно или несколько допущений не выполняются, хотя технически возможно выполнить эти тесты, было бы неправильно интерпретировать результаты и доверять выводам.

Ниже приведены предположения ANOVA, способы их проверки и какие другие тесты существуют, если предположение не выполняется:

Тип переменной: ANOVA требует сочетания одной непрерывной количественной зависимой переменной (которая соответствует измерениям, к которым относится вопрос) и одной качественной независимой переменной (по крайней мере, с 2 уровнями, которые будут определять группы для сравнения).

Независимость: данные, собранные от репрезентативной и случайно выбранной части общей совокупности, должны быть независимыми между группами и внутри каждой группы.

Нормальность:
В случае небольших выборок остатки2 должны приблизительно соответствовать нормальному распределению.

Равенство дисперсий: дисперсии разных групп в популяциях должны быть равными (предположение, называемое однородностью дисперсий, или даже иногда называемое гомоскедастичностью, в отличие от гетероскедастичности, если дисперсии различны в разных группах).

Выбросы: Выброс - это значение или наблюдение, удаленное от других наблюдений. Не должно быть существенных отклонений в разных группах, иначе выводы вашего ANOVA могут быть ошибочными.
