Воспользуйтесь функцией lm() для подгонки простой линейной регрессии с mpg в качестве отклика и horsepower в качестве предиктора для набора данных auto. Примените функцию summary () для вывода результатов на экран.

Чтение файла csv, в качестве аргумента подается название файла

```r
Auto = read.csv('Auto.csv')
Auto
```

Преобразование в числовой формат c помощью функции numeric


```r
Auto$mpg = as.numeric(Auto$mpg)
Auto$horsepower = as.numeric(Auto$horsepower)
```

Выполняет линейную регрессию между столбцом mpg и столбцом horsepower в датафрейме Auto.

Функция lm() используется для построения линейной регрессии. Она принимает два аргумента: зависимую переменную (y) и независимую переменную (x).

В данном случае, зависимая переменная - это mpg, а независимая переменная - horsepower.

Таким образом, функция lm() будет пытаться найти линейную функцию, которая наилучшим образом описывает зависимость между mpg и horsepower.

Аргумент data = Auto указывает на набор данных, который будет использоваться для построения модели.

```r
reg<- lm(mpg ~ horsepower,data = Auto)
```

Функция summary() используется для получения сводки результатов модели линейной регрессии. Она выводит информацию о значениях параметров модели, статистических тестах, t-тестах и других метриках, которые могут быть использованы для оценки качества модели.

Раздел "Residuals": это раздел, в котором отображаются различия между фактическими значениями целевой переменной (mpg) и предсказанными значениями (y-hat) модели. В данном случае, модель содержит 5 наблюдений, из которых 3 были удалены из-за отсутствия значений целевой переменной.

Раздел "Coefficients": это раздел, в котором отображаются значения коэффициентов модели. В данном случае, коэффициент Estimate для переменной horsepower равен -0.1578, а коэффициент Estimate для переменной Intercept равен 39.9359. Коэффициенты Std. Error показывают стандартные ошибки для коэффициентов, которые могут быть использованы для оценки того, насколько значимы эти значения для прогнозирования значений целевой переменной.

Раздел "Signif. codes": это раздел, в котором отображаются коды, которые указывают на уровень значимости статистических тестов, которые были выполнены для проверки гипотезы о нулевой зависимости между переменными.

Раздел "Residual standard error": это раздел, в котором отображается стандартная ошибка для разниц между фактическими значениями целевой переменной и предсказанными значениями.

Раздел "Multiple R-squared": это раздел, в котором отображается коэффициент детерминации (R^2) и коэффициент детерминации, учитывающий количество наблюдений (Adjusted R-squared). Коэффициент детерминации показывает, какую долю дисперсии зависимой переменной можно объяснить с помощью модели линейной регрессии. Коэффициент детерминации, учитывающий количество наблюдений, показывает, какую долю дисперсии зависимой переменной можно объяснить с помощью модели линейной регрессии, учитывая количество наблюдений.

Раздел "F-statistic": это раздел, в котором отображается статистический тест Фишера (F-statistic) и значение p-value для этого теста. Статистический тест Фишера используется для проверки гипотезы о нулевой зависимости между переменными. p-value показывает вероятность получения статистического теста, если он был бы получен случайно. Если p-value меньше уровня значимости (обычно 0.05), то можно сделать вывод о том, что гипотеза о нулевой зависимости между переменными не отвергается.

```r
summary(reg)
```

Результаты регрессии показывают, что коэффициент "Estimate" для переменной "horsepower" равен -0.157845, что означает, что при увеличении мощности двигателя на 1 лошадиная сила, среднее количество миль на 100 км автомобиля уменьшается на 0.157845.

Также регрессия показывает, что коэффициент "t value" для переменной "horsepower" равен -24.49, что означает, что статистический тест, используемый для проверки гипотезы о независимости переменных, не отвергает нулевую гипотезу.

Таким образом, мы можем заключить, что мощность двигателя автомобиля влияет на среднее количество миль на 100 км автомобиля.

# **Задание 2**

```r
set.seed(1) # Инициализация генератора случайных чисел

# Генерация набора данных
x1 = runif(100)                    # генерация 100 случайных значений от 0 до 1
x2 = 0.5*x1 + rnorm(100)/10        # генерация 100 случайных значений с нормальным распределением на основе x1
y = 2 + 2*x1 + 0.3*x2 + rnorm(100) # генерация 100 случайных значений с нормальным распределением
```

* y - это зависимая переменная, которую мы хотим предсказать.
* ~ - это оператор, который указывает на то, что мы хотим построить модель линейной регрессии.
* x1 и x2 - это независимые переменные, которые мы используем для предсказания значений зависимой переменной.

Таким образом, функция lm() пытается найти линейную функцию, которая наилучшим образом описывает зависимость между y и x1, x2.

```r
fun <- lm(y ~ x1+x2)
summary(fun)
```

Новый датафрейм "data" с тремя столбцами: "x1", "x2" и "y".

Далее, функция "attach" используется для присоединения датафрейма "data" к текущей среде. Это означает, что все переменные из датафрейма "data" будут доступны в текущей среде R и могут быть использованы для выполнения анализа данных.

```r
data = data.frame(x1, x2, y)
attach(data)
```

```r
x1i <- seq(0.9*min(x1), 1.1*max(x1), len=100)
x2i <- seq(0.9*min(x2), 1.1*max(x2), len=100)
pre <- predict(fun,data.frame(x1=x1i,x2=x2i),interval="confidence")
```

x1i <- seq(0.9*min(x1), 1.1*max(x1), len=100) - это функция seq(), которая используется для создания массива значений x1i с равными шагами между минимальным и максимальным значениями x1.

Аргументы функции seq():

* 0.9*min(x1) - это начальное значение последовательности.
* 1.1*max(x1) - это конечное значение последовательности.
* len=100 - это количество значений в последовательности.
Таким образом, эта строка кода создает массив значений x1i с шагом в 0.1 между минимальным и максимальным значениями x1.

x2i <- seq(0.9*min(x2), 1.1*max(x2), len=100) - это функция seq(), которая используется для создания массива значений x2i с равными шагами между минимальным и максимальным значениями x2.

Аргументы функции seq():

* 0.9*min(x2) - это начальное значение последовательности.
* 1.1*max(x2) - это конечное значение последовательности.
* len=100 - это количество значений в последовательности.
Таким образом, эта строка кода создает массив значений x2i с шагом в 0.1 между минимальным и максимальным значениями x2.

pre <- predict(fun,data.frame(x1=x1i,x2=x2i),interval="confidence") - это функция predict() в R, которая используется для предсказания значений зависимой переменной на основе модели линейной регрессии, построенной с помощью функции lm().

Аргументы функции predict():

* fun - это результат выполнения функции lm(), который содержит информацию о модели линейной регрессии.
* data.frame(x1=x1i,x2=x2i) - это новые данные, на основе которых нужно предсказать значения. В данном случае, мы используем массивы x1i и x2i для предсказания значений зависимой переменной.
* interval="confidence" - это параметр, который указывает на то, что мы хотим получить интервалы уверенности для предсказанных значений.

Таким образом, эта строка кода использует модель линейной регрессии, построенную с помощью функции lm(), для предсказания значений зависимой переменной для новых значений x1 и x2.

Аргумент interval="confidence" указывает на то, что мы хотим получить интервалы уверенности для предсказанных значений.

```r
plot(x1, y, pch=3)
matplot(x1i, pre, type="l", lty=c(1,2,2), add=TRUE)
```

В третьей строке выводятся результаты моделирования с использованием переменной x2. В этой модели мы предполагаем, что зависимая переменная (y) зависит от двух независимых переменных (x1 и x2).

В четвертой строке выводятся коэффициенты модели. Коэффициенты показывают, насколько сильно каждый аргумент влияет на предсказание модели. В данном случае, коэффициенты модели содержат значение коэффициента перед интерцептором (2.112) и значение коэффициента перед переменной x1 (1.976).

# Для устранения мультиколлинеарности можно использовать методы стандартизации данных и выбросов.

Метод step() используется для удаления столбцов с мультиколлинеарностью. Он позволяет удалить столбцы, которые сильно коррелируют с другими столбцами. Для этого используется коэффициент детерминации (VIF), который определяет, насколько сильно коррелирует столбец с другими столбцами. Если VIF превышает определенный порог, то столбец удаляется.

Метод vif() используется для вычисления коэффициента детерминации (VIF) для каждого столбца в датасете. Если VIF превышает определенный порог, то это может указывать на мультиколлинеарность.
